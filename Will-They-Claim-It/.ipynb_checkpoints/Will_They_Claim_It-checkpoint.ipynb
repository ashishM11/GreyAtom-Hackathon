## (A) Data Pre-Processing and Exploratory Data Analysis

```python
def check_data_missing_columns(dataframe):
    """ Function verify any missing data in dataframe. """    
    data_missing_columns = [col for col in train_df.columns if train_df[col].isnull().sum() > 0]
    if len(data_missing_columns) == 0:
        print("No Data missing from Dataset.")
    else:
        print("Data missing in columns : ",",".join(data_missing_columns))
        
check_data_missing_columns(train_df)
```

```python
def remove_unwanted_features(dataframe,remove_cols):
    dataframe = dataframe.drop(columns=remove_cols,axis=1)
    return dataframe

train_df = remove_unwanted_features(train_df,['ID'])
```

```python
# Since Data is not missing now verifying the columns and its type
def get_numeric_and_non_numeric_columns(dataframe):
    numeric_df = dataframe.select_dtypes(include=np.number)
    non_numeric_df = dataframe.select_dtypes(exclude=np.number)
    return numeric_df,non_numeric_df

numeric_df , non_numeric_df = get_numeric_and_non_numeric_columns(train_df)
```

```python
def check_value_counts(dataframe,cate_cols):
    for col in cate_cols:
        sns.set(style="darkgrid")
        if col == 'Destination':
            sns.set(rc={'figure.figsize':(10,20)})
            ax = sns.countplot(y=col, data=dataframe, orient='h')
        else:
            sns.set(rc={'figure.figsize':(10,10)})
            ax = sns.countplot(x=col,hue='Claim',data=dataframe, orient='h')
        ax.set_xticklabels(ax.get_xticklabels(),rotation=90)
        plt.show()

check_value_counts(train_df,non_numeric_df.columns)
```

```python
# Pair Plot
sns.pairplot(data=train_df,hue='Claim',dropna=True)
plt.show()
```

```python
X = train_df.iloc[:,0:-1]
y = train_df.iloc[:,-1]
print(type(X)," => Shape of train dataframe = ",X.shape)
print(type(y)," => Shape of target series  = ",y.shape)
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=5,stratify=y)
```

```python
def display_target_distribution_plot(target_type,target):
    target_cnt = target.value_counts()
    target_cnt.sort_index().plot(kind = 'bar',color=['skyblue','lime'],figsize=(5,5))
    plt.xlabel('Class')
    plt.ylabel('Data points per Class')
    plt.title('Distribution of target in {} data'.format(target_type))
    plt.grid()
    plt.show()
    for i in range(0,2):
        print('No. of Data points in class {} : {} ({}%)'.format(i,target_cnt[i],round(target_cnt[i]/len(target)*100,2)))
        print('-'*80)

        
train_test_target_list = {
    "Train":y_train,
    "Test":y_test
}
        
for key,target in train_test_target_list.items():
    display_target_distribution_plot(key,target)
```

## (B) Feature Transformation {Scaling and Encoding}

#### {B.1} ENCODING CATEGORICAL COLUMN

```python
num_cols = numeric_df.columns[:-1]
cat_cols = non_numeric_df.columns
print(num_cols)
print(cat_cols)
```

```python
one_hot_enc = OneHotEncoder(handle_unknown='ignore')
```

```python
one_hot_enc_train_df = pd.DataFrame(one_hot_enc.fit_transform(X_train.loc[:,cat_cols]).toarray())
one_hot_enc_test_df =  pd.DataFrame(one_hot_enc.transform(X_test.loc[:,cat_cols]).toarray())
```

```python
print(one_hot_enc_train_df.shape)
print(one_hot_enc_test_df.shape)
```

#### {B.2} SCALING OF NUMERICAL COLUMN

```python
standard_scaler = StandardScaler(with_mean=False)
```

```python
scaled_train_df = pd.DataFrame(standard_scaler.fit_transform(X_train.loc[:,num_cols]),columns=num_cols)
scaled_test_df = pd.DataFrame(standard_scaler.transform(X_test.loc[:,num_cols]),columns=num_cols)
```

```python
print(scaled_train_df.shape)
print(scaled_test_df.shape)
```

### Combining the Transformed Feature into single dataframe.

```python
final_train_df = pd.concat([one_hot_enc_train_df,scaled_train_df],1)
final_test_df = pd.concat([one_hot_enc_test_df,scaled_test_df],1)

print("**  Shape after combining Scaling and Encoding **")
print(" Training Dataframe Shape : ",final_train_df.shape)
print(" Testing Dataframe Shape : ",final_test_df.shape)
```
